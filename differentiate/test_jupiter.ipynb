```python
import numpy as np
import matplotlib.pyplot as plt   

# create mock data
data = np.random.random([100, 100]) * 10
```

```python
print(np.max(data))
print(data.shape)
print(data)
```

```python
plt.figure(figsize=(7, 6))
plt.pcolormesh(data)
plt.colorbar()
```

```python
plt.pcolormesh(data, cmap='inferno')
```

```python
from matplotlib import cm
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
```

```python
new_inferno = cm.get_cmap('seismic', 256)
# visualize with the new_inferno colormaps
plt.pcolormesh(data, cmap = new_inferno)
plt.colorbar()
```

Change range of colour

```python
# modified hsv in 256 color class
hsv_modified = cm.get_cmap('jet', 256)
# create new hsv colormaps in range of 0.3 (green) to 0.7 (blue)
newcmp = ListedColormap(hsv_modified(np.linspace(0.3, 0.7, 256)))# show figure
plt.figure(figsize=(7, 6))
plt.pcolormesh(data, cmap = newcmp)
plt.colorbar()
```
